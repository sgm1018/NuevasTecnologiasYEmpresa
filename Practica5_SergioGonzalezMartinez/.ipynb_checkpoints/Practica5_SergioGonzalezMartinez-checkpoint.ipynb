```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
```

Red de familias Medici vs Pazzi

```python
grafo=nx.florentine_families_graph()
nx.draw(grafo,with_labels=True)
plt.show()
```

Utilizaremos la medida de centralidad por Cercania, de esta manera podremos saber cuales son los nodos o familias son las mejores para monitorear el flujo de la informacion, por tanto, esto significa tener mas influencia entre las familias, ya que podemos saber si mediante nuestros enlaces tenemos la capacidad de llegar a cualquier punto de la red mas rapidamente.

```python
print(nx.betweenness_centrality(grafo, k=None, normalized=True, weight=None, endpoints=False, seed=None))
```

```python
#Sacamos el rango
rank=nx.pagerank(grafo, alpha=0.85, personalization=None, max_iter=100, tol=1e-06, nstart=None, weight='weight', dangling=None)
print(nx.pagerank(grafo, alpha=0.85, personalization=None, max_iter=100, tol=1e-06, nstart=None, weight='weight', dangling=None))
```

```python
listaRangosOrdenada=[]
#Funcion que complementa la funcion Key del sort de las listas
def cogerSegundo(elemento):
    return elemento[1]
#Ordenamos los nodos
for (x,y) in rank.items():
    listaRangosOrdenada.append((x,y))
listaRangosOrdenada.sort(key=cogerSegundo)
listaRangosOrdenada.reverse()
print(listaRangosOrdenada)
```

```python
NodosYcolores=[]
nodos=[]
color=1
#Asignamos colores a nuestros nodos de manera ordenada.
for (node,value) in listaRangosOrdenada:
    NodosYcolores.append((node,color))
    color-=0.01
colores_map=dict(NodosYcolores)
valores_colors=[colores_map.get(node) for node in grafo.nodes()]
print("NUESTRO NODOS ORDENADOS")
print(grafo.nodes())
print("NUESTRO MAPA DE COLORES, este le usaremos para darles preferencia de color al dibujar el mapa")
print(valores_colors)
print("NUESTROS NODOS Y COLORES")
print(colores_map)  
     
```

```python
nx.draw(grafo, cmap=plt.cm.YlGn, node_color=valores_colors, with_labels=True, font_color='black', node_size=500)
```

Cuanto mas oscuro, mas rango tiene.
