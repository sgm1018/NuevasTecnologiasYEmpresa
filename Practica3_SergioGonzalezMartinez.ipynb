```python
#Alumno: Sergio González Martínez
#N - numero de nodos
#P - par de nodos(i,j)
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
def redAleatoria(N,P):
    g=nx.Graph()
    g.add_nodes_from(range(1,N))
    for x in list(g.nodes()):
        for y in list(g.nodes()):
            aleatorio = random.random()
            if (aleatorio<P):
                g.add_edge(x,y)
    return g
                
    
```

```python
#Ejercicio 1 (N=100, P=0.01)
red=redAleatoria(100,0.01)
mAdyacencia=nx.adjacency_matrix(red, nodelist=None, dtype=None, weight='weight').todense()
plt.matshow(mAdyacencia)
print("El numero de enlaces es: ",red.number_of_edges())
print("Grado medio de la red: ",np.mean([k[1] for k in red.degree()]))
```

```python
#Layout circular
nx.draw(red,pos=nx.circular_layout(red))
```

```python
#Ejercicio 2 (N=100 P=0.1)
red2=redAleatoria(100,0.1)
print("El numero de enlaces es: ",red2.number_of_edges())
print("Grado medio de la red: ",np.mean([k[1] for k in red2.degree()]))
camino=nx.shortest_path(red2,6,36) #Camino mas corto
#Mostrar camino.
caminoLinks=list(zip(camino,camino[1:]))
posicion=nx.layout.kamada_kawai_layout(red2)
nx.draw(red2,posicion,with_labels=True,node_color='b',alpha=0.5)
nx.draw_networkx_nodes(red2,posicion,nodelist=camino,node_color='r',alpha=0.5)
nx.draw_networkx_edges(red2,posicion,edgelist=caminoLinks,edge_color='r',width=2)
```
